Un Hombre desea saber cuanto dinero se genera por concepto de intereses sobre la cantidad que tiene en inversion en el banco. El decidira reinvertir los intereses siempre y cuando estos excedan a $7000, y en ese caso desea saber cuanto dinero tendra finalmente en su cuenta.

```python
cap = float(input("capital a invertir, $: "))
tasa_int = float(input("Interes, %: "))
intereses = (cap * tasa_int)/100
if(intereses >7000):
  cap - cap + intereses
print("La cuenta actual es de $:", cap)
```

En una tienda se ha establecido la siguiente oferta: por compras menores a 50 pesos se hara un descuento del 8% pero para compras a partir de $50 el descuento es de 10%. Se pide ingresar la cantidad y el precio del producto que se compra y determinar cuanto se descontará y cuanto se cobrará.

```python
cantidad = int(input("ingresar cantidad de producto: "))
precio = float(input("Ingrsar el precio de producto: "))

pago = cantidad + precio

if (pago < 50):
  descuento = pago * 0.8 
  pago = pago - descuento
else:
  descuento = pago * 0.1
  pago = pago - descuento
  
print("El descuento es: ",round(descuento,2))
print("El pago a realizar es: ",round(pago,2))
```
